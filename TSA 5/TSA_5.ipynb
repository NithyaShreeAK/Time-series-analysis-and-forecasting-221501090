```python
# 📦 Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# 📥 Load Dataset
df = pd.read_csv('/content/MLTempDataset.csv')
df['Datetime'] = pd.to_datetime(df['Datetime'])
df.set_index('Datetime', inplace=True)

# 📊 Daily Aggregation
daily_data = df['DAYTON_MW'].resample('D').mean()

# 🔁 Trend Estimation using 7-day Rolling Mean
trend = daily_data.rolling(window=7).mean()

# 📈 Visualization
plt.figure(figsize=(14, 6))
plt.plot(daily_data, label='Original', color='blue')
plt.plot(trend, label='Trend (7-day Rolling Mean)', color='red')
plt.title('Trend Estimation in Time Series Data')
plt.xlabel('Date')
plt.ylabel('DAYTON_MW')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
