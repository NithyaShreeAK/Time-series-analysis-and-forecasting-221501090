```python
# üì¶ Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# üì• Load Dataset
df = pd.read_csv('/content/MLTempDataset.csv')
df['Datetime'] = pd.to_datetime(df['Datetime'])
df.set_index('Datetime', inplace=True)

# üìä Daily Aggregation
daily_data = df['DAYTON_MW'].resample('D').mean()

# üîÅ Moving Average Smoothing (3-day)
smoothed = daily_data.rolling(window=3).mean()

# üìà Visualization
plt.figure(figsize=(14, 6))
plt.plot(daily_data, label='Original Data', color='gray')
plt.plot(smoothed, label='3-day Moving Average', color='green')
plt.title('Moving Average Smoothing for Time Series Forecasting')
plt.xlabel('Date')
plt.ylabel('DAYTON_MW')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
